```python
class Point:
    def __init__(self, a, b):
        self.a = a
        self.b = b
            
    def show(self):
        print('(', self.a, ',', self.b, ')')

a = Point(1, 2)
a.show()

b = Point(3, 4)
b.show()

a = a+b
a.show() 
```

```python
#2021년 5월 1일부터 31일까지의 기온데이터
#일교차의 합이 가장 큰 1주일이 언제부터 언제인지?

#1일부터 하루씩 이동하며 7일 단위로 일교차의 합 계산 
#합이 가장 큰 일주일 시작 날짜 & 마지말 날짜 출력

import csv

f = open("seoul.csv") 
data = csv.reader(f)
header = next(data)

max_temp = -100
week_temp = 0

i = 0
day = 1

for row in data:
    if row[-1] == '':
        row[-1] == 100
    row[-1] = float(row[-1])
    
    if i < 7:
        week_temp += float(row[4])-float(row[3])
        
    elif i == 7:
        if max_temp < week_temp:
            max_week = day - 7 #일주일의 시작은 7일 전이므로 
            max_temp = week_temp
        
        i = 0
        week_temp = 0       
        week_temp += float(row[4])-float(row[3])

    i += 1
    day += 1
        
f.close()

print("Max Week = 5/", max_week, "~ 5/", max_week + 6)

        
```
