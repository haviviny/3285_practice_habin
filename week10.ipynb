```python
num = list(map(int,input().split()))
max = 0

for i in num :
    if max == 0 or max < i:
        max = i
        
print ("Max = ", max)
```

```python
#encryption 
str = input("input of encryption: ")
str = str.split()
str.reverse()
str = ' '.join(str)
en = ""

for i in range(len(str)) :
    c = str[i]
    if (c != ' ') :
        ac = ord(c)
        ac += 1
        c = chr(ac)
    en += c
    
print("result of enrcryption: ", en)

#decrptyion 
str = input("input of decryption: ")
str = str.split()
str.reverse()
str = ' '.join(str)
de = ""

for i in range(len(str)) :
    c = str[i]
    if (c != ' ') :
        ac = ord(c)
        ac -= 1
        c = chr(ac)
    de += c
    
print("result of decryption: ", de)
```

```python
book = {}

for i in range(3) :
    name = input('name: ').title()
    phone = input('phone: ')
    book[name] = phone
    
print("search name or phone?")
print("1. name")
print("2. phone")

while True:
    menu = input("menu: ")
    temp_book = dict(map(reversed, book.items()))
    
    if menu == '1':
        name = input('name?: ').title()
        phone = book.get(name)
        print('phone number of', f'{name}', 'is', phone)
        
    elif menu == '2': 
        phone = input('phone?: ').title()
        name = temp_book.get(phone)
        print(f'{phone}', 'is phone number of', name)
```
