```python
import csv
import numpy as np
import matplotlib.pyplot as plt 

f = open('seoul_age.csv')
data = csv.reader(f)
next(data)
data = list(data)

min = 1
near = ''
near_dis = 0

max = 0
far = ''
far_dis = 0

for row in data:
    if ' ' in row[1]:
        gwangjin = np.array(row[4:], dtype = int) / int(row[3])
        
for row in data:
    other = np.array(row[4:], dtype = int) / int(row[3])
    d = np.sum((gwangjin - other) **2)
    
    if d < min and ' ' not in row[1]: 
        min = d
        near = row[1]
        near_dis = other    
    elif d > max and ' ' not in row[1]:
        max = d
        far = row[1]
        far_dis = other

plt.style.use('ggplot')
plt.figure(figsize = (8,5), dpi = 300)
plt.rc('font', family='Malgun Gothic')
plt.title('        ')
plt.plot(gwangjin, label = ' ')
plt.plot(near_dis, label = '  : ' + near)
plt.plot(far_dis, label='  : ' + far)
plt.legend()
plt.show()
```
