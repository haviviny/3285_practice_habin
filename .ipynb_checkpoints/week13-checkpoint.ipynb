```python
import matplotlib.pyplot as plt 
import csv

f = open('seoul_10.csv')
data = csv.reader(f)
next(data)
max_temp = []
min_temp = []

for row in data :
    year = row[0].split('-')[0]
    month = row[0].split('-')[1]
    day = row[0].split('-')[2]
    
    if row[-1] != '':
        if month == '06' and day == '25':
            max_temp.append(float(row[-1]))
            min_temp.append(float(row[-2]))

print("6월 25일의 서울 지역 최고 기온과 최저 기온 (2010 ~ 2019)")
plt.title("Temperature of My Birthday Graph")
plt.plot(max_temp)
plt.plot(min_temp)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import csv

f = open('seoul_10.csv')
data = csv.reader(f)
next(data)

jan = []
aug = []


for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

print("서울 지역의 1월과 8월 최고 기온 데이터 (2010 ~ 2019)")
plt.title("The highest temperatures in January and August in Seoul.")
plt.hist(jan, bins=100, color='b')
plt.hist(aug, bins=100, color='r')
plt.show()

```
