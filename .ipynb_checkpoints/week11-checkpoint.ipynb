```python
#실습 6: cp 명령어 구현
#cp file1.txt file2.txt 입력

userinput = input()
c = userinput.split()

if(c[0] == 'cp'):
    f1 = open (c[1], 'r')
    data = f1.read()
    
    f2 = open (c[2], 'w')
    f2.write(data)
    
    f1.close()
    f2.close()
```

```python
#실습 7: wc 명령어 구현
#wc file1.txt

userinput = input()
c = userinput.split()

line_count = 0
word_count = 0

if(c[0] == 'wc'):
    f = open (c[1], 'r')

    for line in f:
        line_count += 1
        word_count += len(line.split())
        
    f.close()
    
    print ("Number of lines: %d" %line_count)
    print ("Number of Words: %d" %word_count)
```

```python
#실습 8: 사칙 연산 클래스

class Calc:
    def sum(self, a, b):
        result = a + b
        print ("%s + %s = %s 입니다." %(a, b, result))
        
    def sub(self, a, b):
        result = a - b
        print ("%s - %s = %s 입니다." %(a, b, result))
        
    def mul(self, a, b):
        result = a * b
        print ("%s * %s = %s 입니다." %(a, b, result))
        
    def div(self, a, b):
        result = a / b
        print ("%s / %s = %s 입니다." %(a, b, result))
        
calc = Calc()
calc.sum(1, 2)
calc.sub(5, 1)
calc.mul(2, 3)
calc.div(6, 2)
```
