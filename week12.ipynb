```python
import csv

#일교차
def dtr_day(): 
    f = open('seoul.csv')
    data = csv.reader(f)
    header = next(data)

    dtr_temp = -100
    dtr_row = ''
        
    for row in data:
        dtr = float(row[4])-float(row[3])
        
        if row[-1] == '' :
            row[-1] == 100
        row[-1] = float(row[-1])
        
        if dtr_temp < float(row[4])-float(row[3]):
            dtr_date = row[0] 
            dtr_temp = dtr
            
    f.close()
    
    return dtr_date
    
#최저 기온
def min_day():
    f = open('seoul.csv')
    data = csv.reader(f)
    header = next(data)
    
    min_temp = 100
    min_date = ''

    for row in data:
        if row[-1] == '' :
            row[-1] == 100
        row[-1] = float(row[-1])
        
        if min_temp > row [-1]:
            min_date = row[0]
            min_temp = row[-1]

    f.close()
    
    return min_date


#최고 기온 
def max_day():
    f = open('seoul.csv')
    data = csv.reader(f)
    header = next(data)

    max_temp = -100
    max_date = ''

    for row in data:
        if row[-1] == '' :
            row[-1] == -100
        row[-1] = float(row[-1])
        
        if max_temp < row [-1]:
            max_date = row[0]
            max_temp = row[-1]
            
    f.close()
    
    return max_date

print('일교차가 가장 큰 날\n', dtr_day())
print('최저기온이 가장 낮은 날\n', min_day())
print('최고기온이 가장 높은 날\n', max_day())
```
